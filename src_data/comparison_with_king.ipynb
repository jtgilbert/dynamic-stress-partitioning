```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('transport_data_cleaned_king.csv')
```

```python
df['Wi_star'] = (1.65*9.81*df['qsi']*1000/86400)/((df['Fi']*100)*(9.81*df['h']*df['S'])**1.5)
```

```python
df_foj = df[df['Stream']=='Fourth of July Creek']
df_cs = df[df['Stream']=='Cat Spur Creek']
df_l = df[df['Stream']=='Lochsa River']
df_lc = df[df['Stream']=='Lolo Creek']
df_rr = df[df['Stream']=='Red River']
df_nfcw = df[df['Stream']=='North Fork Clearwater']
df_sfrr = df[df['Stream']=='South Fork Red River']
df_vc = df[df['Stream']=='Valley Creek']
df_tc = df[df['Stream']=='Trapper Creek']
```

```python
x1 = np.arange(0.7, 2.01, 0.01)
y1 = 0.0008*x1**7.5
x2 = np.arange(1.99, 10, 0.01)
y2 = 14*(1-(1.11/x2**0.8))**4.5
```

```python
fig, ax = plt.subplots(3,3, figsize=(12,16))
fig.subplots_adjust(wspace=0.35)
pts1 = ax[0,0].scatter(df_foj['gs_ratio'], df_foj['Wi_star'], c=df_foj['D'])
ax[0,0].plot(x1, y1, c='r', linewidth=2)
ax[0,0].plot(x2, y2, c='r', linewidth=2)
ax[0,0].set_xscale('log')
ax[0,0].set_yscale('log')
ax[0,0].set_title('Fourth of July')
plt.colorbar(pts1, ax=ax[0,0])

pts2 = ax[0,1].scatter(df_cs['gs_ratio'], df_cs['Wi_star'], c=df_cs['D'])
ax[0,1].plot(x1, y1, c='r', linewidth=2)
ax[0,1].plot(x2, y2, c='r', linewidth=2)
ax[0,1].set_xscale('log')
ax[0,1].set_yscale('log')
ax[0,1].set_title('Cat Spur')
plt.colorbar(pts2, ax=ax[0,1])

pts3 = ax[0,2].scatter(df_l['gs_ratio'], df_l['Wi_star'], c=df_l['D'])
ax[0,2].plot(x1, y1, c='r', linewidth=2)
ax[0,2].plot(x2, y2, c='r', linewidth=2)
ax[0,2].set_xscale('log')
ax[0,2].set_yscale('log')
ax[0,2].set_title('Lochsa')
plt.colorbar(pts3, ax=ax[0,2])

pts4 = ax[1,0].scatter(df_lc['gs_ratio'], df_lc['Wi_star'], c=df_lc['D'])
ax[1,0].plot(x1, y1, c='r', linewidth=2)
ax[1,0].plot(x2, y2, c='r', linewidth=2)
ax[1,0].set_xscale('log')
ax[1,0].set_yscale('log')
ax[1,0].set_title('Lolo')
plt.colorbar(pts4, ax=ax[1,0])

pts5 = ax[1,1].scatter(df_rr['gs_ratio'], df_rr['Wi_star'], c=df_rr['D'])
ax[1,1].plot(x1, y1, c='r', linewidth=2)
ax[1,1].plot(x2, y2, c='r', linewidth=2)
ax[1,1].set_xscale('log')
ax[1,1].set_yscale('log')
ax[1,1].set_title('Red')
plt.colorbar(pts5, ax=ax[1,1])

pts6 = ax[1,2].scatter(df_nfcw['gs_ratio'], df_nfcw['Wi_star'], c=df_nfcw['D'])
ax[1,2].plot(x1, y1, c='r', linewidth=2)
ax[1,2].plot(x2, y2, c='r', linewidth=2)
ax[1,2].set_xscale('log')
ax[1,2].set_yscale('log')
ax[1,2].set_title('North Fork Clearwater')
plt.colorbar(pts6, ax=ax[1,2])

pts7 = ax[2,0].scatter(df_sfrr['gs_ratio'], df_sfrr['Wi_star'], c=df_sfrr['D'])
ax[2,0].plot(x1, y1, c='r', linewidth=2)
ax[2,0].plot(x2, y2, c='r', linewidth=2)
ax[2,0].set_xscale('log')
ax[2,0].set_yscale('log')
ax[2,0].set_title('South Fork Red')
plt.colorbar(pts7, ax=ax[2,0])

pts8 = ax[2,1].scatter(df_vc['gs_ratio'], df_vc['Wi_star'], c=df_vc['D'])
ax[2,1].plot(x1, y1, c='r', linewidth=2)
ax[2,1].plot(x2, y2, c='r', linewidth=2)
ax[2,1].set_xscale('log')
ax[2,1].set_yscale('log')
ax[2,1].set_title('Valley')
plt.colorbar(pts8, ax=ax[2,1])

pts9 = ax[2,2].scatter(df_tc['gs_ratio'], df_tc['Wi_star'], c=df_tc['D'])
ax[2,2].plot(x1, y1, c='r', linewidth=2)
ax[2,2].plot(x2, y2, c='r', linewidth=2)
ax[2,2].set_xscale('log')
ax[2,2].set_yscale('log')
ax[2,2].set_title('Trapper')
plt.colorbar(pts9, ax=ax[2,2])

fig.text(0.5, 0.09, r'$\tau_g/\tau_{gc}$', ha='center', fontsize=20)
fig.text(0.04, 0.5, r'$W_i^*$', fontsize=20)
fig.text(0.92, 0.5, 'D (m)', fontsize=20)
#plt.savefig('comparison_with_king.png', dpi=150)
plt.show()
```
